```python
import sklearn.datasets
import pandas as pd

#load dataset as dataframe
cancer =sklearn.datasets.load_breast_cancer(return_X_y=True, as_frame=True)
type(cancer[0]), type( cancer[1])

cancer[1]

#convert cancer dataset to dataframe
a= cancer[0]
a['typeofcancer']=cancer[1]

a.shape
a['typeofcancer']

cancer[1]
```

```python
df = pd.DataFrame
df =a.iloc[[17,18,19,20,21], [0,2,3,30]]
df.head()
```

```python
#create new dataframe containing only malignant cases
malignant = a[a['typeofcancer']==0]

#create new dataframe containing only benign cases
benign = a[a['typeofcancer']==1]
```

```python
malignant
```

```python
benign
```

```python
import matplotlib.pyplot as plt


fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,4))

#plot subplot1
ax1.hist(malignant['mean radius'],color=(0.2, 0.4, 0.6),  edgecolor='black', linewidth=1.2, alpha=0.7,label='malignant')
ax1.hist(benign['mean radius'],color=(1, 0.8,0.6),  edgecolor='red', linewidth=1.2, alpha=0.7, label='Benign')

#add subplot 1 details
ax1.set_xlabel('mean radius')
ax1.set_ylabel('Frequency')
ax1.legend()

#plot subplot 2
ax2.scatter(malignant['mean perimeter'], malignant['mean radius'], label ='c0(M)')
ax2.scatter(benign['mean perimeter'], benign['mean radius'], label ='c1(B)')

#add subplot 2 details
ax2.set_xlabel('mean perimeter')
ax2.set_ylabel('mean radius')
ax2.legend()

#plot subplot 3
ax3.scatter(malignant['mean area'], malignant['mean radius'], label ='c0(M)')
ax3.scatter(benign['mean area'], benign['mean radius'], label ='c1(B)')

#add subplot 3 details
ax3.set_xlabel('mean area')
ax3.set_ylabel('mean radius')
ax3.legend()
```

```python
import matplotlib.pyplot as plt


fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize=(18,3))

#plot subplot1
ax1.hist(malignant['mean radius'],color=(0.2, 0.4, 0.6),  edgecolor='black', linewidth=1.2, alpha=0.7,label='malignant')
ax1.hist(benign['mean radius'],color=(1, 0.8,0.6),  edgecolor='red', linewidth=1.2, alpha=0.7, label='Benign')

#add subplot 1 details
ax1.set_xlabel('mean radius')
ax1.set_ylabel('Frequency')
ax1.legend()

#plot subplot 2
ax2.scatter(malignant['mean concavity'], malignant['mean radius'], label ='c0(M)')
ax2.scatter(benign['mean concavity'], benign['mean radius'], label ='c1(B)')

#add subplot 2 details
ax2.set_xlabel('mean concavity')
ax2.set_ylabel('mean radius')
ax2.legend()

#plot subplot 3
ax3.scatter(malignant['mean concave points'], malignant['mean radius'], label ='c0(M)')
ax3.scatter(benign['mean concave points'], benign['mean radius'], label ='c1(B)')

#add subplot 3 details
ax3.set_xlabel('mean concave points')
ax3.set_ylabel('mean radius')
ax3.legend()

#plot subplot 4
ax4.scatter(malignant['mean symmetry'], malignant['mean radius'], label ='c0(M)')
ax4.scatter(benign['mean symmetry'], benign['mean radius'], label ='c1(B)')

#add subplot 4 details
ax4.set_xlabel('mean symmetry')
ax4.set_ylabel('mean radius')
ax4.legend()
```
